Web Scraping Tests
======
## Goals:
scrape tabbycat webpages
## Using:
[BeautifulSoup](https://beautiful-soup-4.readthedocs.io/en/latest/index.html?highlight=website)<br>
[Requests](https://requests.readthedocs.io/en/latest/user/quickstart/)

```python
from bs4 import BeautifulSoup
import requests
```

```python
link = 'https://hhiv2021.calicotab.com/HHIV2021/'
r = requests.get(link)
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
tabs = soup.find_all("a", "list-group-item-action")

print(tabs[0])
```

```python
links = [tab.get('href') for tab in tabs]
links
```

```python
stats = list(filter(lambda x: "statistics" in x, links))[0]
stats
```

```python
new_link = link+"motions/statistics/"
r = requests.get(new_link)
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
motions = soup.find_all("h4", "mb-3 mt-1")
motions = [motion.get_text().strip() for motion in motions]
motions
```
